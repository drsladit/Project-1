```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# reading CSV data and creating visualisation
x = pd.read_csv('Top 30days Installs')['app_name']
y = pd.read_csv('Top 30days Installs')['app_installs_30d']
```

```python
# Using Matplotlib in represting

fig_30days = plt.figure()
axes = fig_30days.add_axes([0.5, 0.5, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)
axes.bar(x,y)
```

```python
fig_30days.savefig("Top_30days_apps.png", dpi=200)
```
