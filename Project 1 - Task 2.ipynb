```python
import pandas as pd
import numpy as np
```

```python
app_data = pd.read_json("package_info.json")
```

```python
app_data
```

```python
# Pulling and adding 3 columns to the app_data Dataframe
app_installs_30d = []
app_installs_90d = []
app_installs_365d = []
for x in app_data['app_installs']:
    app_installs_30d.append(x['30days'])
    app_installs_90d.append(x['90days'])
    app_installs_365d.append(x['365days'])
app_data['app_installs_30d'] = app_installs_30d
app_data['app_installs_90d'] = app_installs_90d
app_data['app_installs_365d'] = app_installs_365d
```

```python
app_data # app_data dataframe with new columns added
```

```python
# pull only top 10 values from the app_data for 30days, 90 days, and 365days install. 
# Create new Dataframe
#df_30days_installs_10apps

df_30days_installs_10apps = app_data.sort_values('app_installs_30d', ascending=False).iloc[0:10]
df_90days_installs_10apps = app_data.sort_values('app_installs_90d', ascending=False).iloc[0:10]
df_365days_installs_10apps = app_data.sort_values('app_installs_365d', ascending=False).iloc[0:10]
```

```python
df_30days_installs_10apps.to_csv('Top 30days Installs', index=False)
```

```python
df_90days_installs_10apps.to_csv('Top 90days Installs', index=False)
```

```python
df_365days_installs_10apps.to_csv('Top 365days Installs', index=False)
```
